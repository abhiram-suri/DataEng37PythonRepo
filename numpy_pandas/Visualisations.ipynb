```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Seaborn

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal")

plt.title('fMRI Signal Across Time')
plt.xlabel("Time Point")
plt.ylabel("Percent Signal Change")
```

```python
sns.lineplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    palette=["teal", "indigo"]
)
```

```python
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    col="region",
    kind="line"
)
```

```python
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="region",
    col="event",
    kind="line"
)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
plt.scatter(iris['petal_length'], iris['petal_width'], marker='+')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
vers = iris[iris['species'] == 'versicolor']
virg = iris[iris['species'] == 'virginica']
seto = iris[iris['species'] == 'setosa']

plt.scatter(vers.petal_length, vers.petal_width, marker = 'o')
plt.scatter(virg.petal_length, virg.petal_width, marker = '+')
plt.scatter(seto.petal_length, seto.petal_width, marker = 'x')

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend(labels=["Versicolor", "Virginica", "Setosa"])
plt.suptitle("Petal Length vs Width")
plt.title("For Different Species of Iris")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
boarded = titanic.groupby('class').pclass.count()
boarded
```

```python
import squarify
```

```python
c = ["xkcd:blue", "xkcd:teal","xkcd:pink"]
squarify.plot(boarded, label=boarded.index, color=c)
plt.axis('off')
plt.title("Boarded")
plt.show()
```

```python
c = ["xkcd:blue", "xkcd:teal","xkcd:pink"]
survived = titanic[titanic["alive"] == "yes"].groupby('class').pclass.count()
squarify.plot(survived, label=survived.index, color=c)
plt.axis('off')
plt.title("Survived")
plt.show()
```
