```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
leagues = pd.read_csv("2021-2022 Football Team Stats.csv")
leagues
```

```python
ENG = leagues[leagues['Country'] == 'ENG']
ESP = leagues[leagues['Country'] == 'ESP'].reset_index()
GER = leagues[leagues['Country'] == 'GER'].reset_index()
ITA = leagues[leagues['Country'] == 'ITA'].reset_index()
FRA = leagues[leagues['Country'] == 'FRA'].reset_index()
```

```python
TOT_Att = leagues[['Country', 'Attendance']].groupby('Country').mean().reset_index()
TOT_Att
```

```python
sns.barplot(data=TOT_Att, x="Country", y="Attendance", color="teal")
plt.xlabel("Country")
plt.ylabel("Attendance")
plt.show()
```

```python
TOT_Pts = leagues[['Country', 'Pts/G']].groupby('Country').mean().reset_index()
TOT_Pts
```

```python
sns.barplot(data=TOT_Pts, x="Country", y="Pts/G", color="teal")
plt.xlabel("Country")
plt.ylabel("Points Per Game")
plt.show()
```

```python
TOT_GF = leagues[['Country', 'GF']].groupby('Country').mean().reset_index()
TOT_GF
```

```python
sns.barplot(data=TOT_GF, x="Country", y="GF", color="teal")
plt.xlabel("Country")
plt.ylabel("Goals For")
plt.show()
```

```python
sns.lineplot(data=leagues, x="LgRk", y="Pts/G", hue="Country")
plt.xlabel("League Rank")
plt.ylabel("Points Per Game")
plt.show()
```

```python
sns.lineplot(data=leagues, x="LgRk", y="xGD", hue="Country")
plt.xlabel("League Rank")
plt.ylabel("Expected Goal Difference")
plt.show()
```

```python
sns.lineplot(data=leagues, x="LgRk", y="Attendance", hue="Country")
plt.xlabel("League Rank")
plt.show()
```
